```python
import pandas as pd
from google.colab import files
```

```python
designation_file = pd.read_csv('100K_search_sheet.csv')
```

```python
search_terms = pd.read_csv('Apollo_Div_N_grams_Sorted_Count(Top 20).csv')
```

```python
search_terms.head()
```

```python
lst = designation_file.columns
lst
```

```python
cat = lst[0]
designation_file[cat]
col = designation_file[cat]
```

```python
search_terms_search_list = {}
for i in col[4:]:
  try:
    cl = i.split(" ")
    for j in cl:
      if j.lower() not in search_terms_search_list:
        search_terms_search_list[j.lower()] = [i]
      else:
        search_terms_search_list[j.lower()].append(i)
  except:
    print(i)
```

```python
search_terms_search_list
```

```python
designation = {}
for i in col[4:]:
  try:
    designation[i.lower()] = [0]
  except:
    print(i)
```

```python
designation
```

```python
search_terms.columns
```

```python
division = {}
for i in search_terms.columns:
  if '.' not in i:
    division[i.lower()] = [0]
print(division, len(division))
```

```python
function = {}
for i in search_terms.columns[1:]:
  if search_terms[i][1].lower() not in function:
    function[search_terms[i][1].lower()] = [0]

print(len(function), function)
```

```python
search_terms.head()
```

```python
sss = search_terms_search_list.keys()
sss
```

```python
search_terms.columns
```

```python
try:
    for i in search_terms.columns[1:]:
        try:
          cur_div = i.lower()
          cor_col = search_terms[i]

          if '.' in cur_div:
              ind = cur_div.index('.')
              cur_div = cur_div[:ind]
        except:
          continue
        try:
          cur_function = cor_col[0].lower()
        except:
          continue
        # we got function and division till here.

        for cur_search_term in cor_col[1:]:
            try:
              cur_search_term = cur_search_term.lower()
              cur_search_term = cur_search_term.split(" ")
            except:
              continue
            ans = []
            for des in cur_search_term:
                for item in sss:
                    if des in item:
                        if len(ans):
                            ans = list(set(ans) & set(search_terms_search_list[item]))
                        else:
                            ans = search_terms_search_list[item]
            # we got deginationss till here
            for j in ans:
                # print(ans)
                try:
                    if cur_div not in designation[j]:
                        designation[j][0] += 1
                        designation[j].append(cur_div)

                    if j not in division[cur_div]:
                        division[cur_div][0] += 1
                        division[cur_div].append(j)

                    if j not in function[cur_function]:
                        function[cur_function][0] += 1
                        function[cur_function].append(j)
                except Exception as e:
                    print(e)
except Exception as e:
    print(e)
```
