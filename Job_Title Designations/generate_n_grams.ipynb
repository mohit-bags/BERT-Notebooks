```python
def split_in_csv(word, no):
    word = str(word)
    k = word.split()
    n_gram = list(ngrams(k, n=no))
    
    return n_gram
```

```python
def get_uni_gram(word):
    word = str(word)
    k = word.split()
    
    return k
```

```python
df['uni-gram'] = df.apply(lambda x: get_uni_gram(x["designation"]), axis = 1)
```

```python
df['bi-gram'] = df.apply(lambda x: split_in_csv(x["designation"], 2), axis = 1)
df['tri-gram'] = df.apply(lambda x: split_in_csv(x["designation"], 3), axis = 1)
```

```python
dict_uni={}
for i in df['uni-gram']:
    for j in i:
        if j not in dict_uni:
            dict_uni[j]=1
        else:
            dict_uni[j]+=1
```

```python
dict_uni
```

```python
dict_bi={}
for i in df['bi-gram']:
    i=list(map(" ".join, i))
    for j in i:
        if j not in dict_bi:
            dict_bi[j]=1
        else:
            dict_bi[j]+=1
```

```python
dict_bi
```

```python
dict_tri={}
for i in df['tri-gram']:
    i=list(map(" ".join, i))
    for j in i:
        if j not in dict_tri:
            dict_tri[j]=1
        else:
            dict_tri[j]+=1
```

```python
dict_tri
```

```python
df_uni = pd.DataFrame.from_dict(dict_uni,orient='index')
df_uni=df_uni.reset_index()
df_uni.columns=['Word','Count']
df_uni=df_uni.sort_values(by='Count', ascending = False)
df_uni=df_uni.reset_index(drop=True)
df_uni
```

```python
df_bi = pd.DataFrame.from_dict(dict_bi,orient='index')
df_bi=df_bi.reset_index()
df_bi.columns=['Word','Count']
df_bi=df_bi.sort_values(by='Count', ascending = False)
df_bi=df_bi.reset_index(drop=True)
df_bi
```

```python
df_tri = pd.DataFrame.from_dict(dict_tri,orient='index')
df_tri=df_tri.reset_index()
df_tri.columns=['Word','Count']
df_tri=df_tri.sort_values(by='Count', ascending = False)
df_tri=df_tri.reset_index(drop=True)
df_tri
```
