```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from simpletransformers.ner import NERModel,NERArgs

data1=data1.replace(r'^\s*$', np.nan, regex=True)
data2=data2.replace(r'^\s*$', np.nan, regex=True)
data3=data3.replace(r'^\s*$', np.nan, regex=True)
data4=data4.replace(r'^\s*$', np.nan, regex=True)
data5=data5.replace(r'^\s*$', np.nan, regex=True)
```

```python
data=pd.concat([data1,data2,data3,data4])  ######
data=data.reset_index(drop=True)

test_data=data5   ######

data["Sentence #"] = LabelEncoder().fit_transform(data["Sentence #"] )
test_data["Sentence #"] = LabelEncoder().fit_transform(test_data["Sentence #"] )
data.rename(columns={"Sentence #":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)
test_data.rename(columns={"Sentence #":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)

x_train= data[["sentence_id","words"]]
y_train =data["labels"]
x_test= test_data[["sentence_id","words"]]
y_test =test_data["labels"]


# y_train.value_counts()

#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})


```

Fine Tuning on our Data

```python
label = data["labels"].unique().tolist()
args = NERArgs()
args.num_train_epochs =3
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
# BERT_Pred_2000
model = NERModel('bert', 'bert-base-uncased',labels=label,args =args)
```

```python
#training model on our data
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
result, model_outputs, preds_list = model.eval_model(test_data)
result #after fine tuning on test data
```
