```python
import nltk
import re
from bs4 import BeautifulSoup

nltk.download("punkt")

REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')

def clean_text(text):
    """
        text: a string
        
        return: modified initial string
    """
    text = BeautifulSoup(text, "lxml").text # HTML decoding
    text = text.lower() # lowercase text
    text = REPLACE_BY_SPACE_RE.sub(' ', text) # replace REPLACE_BY_SPACE_RE symbols by space in text
    text = BAD_SYMBOLS_RE.sub('', text) # delete symbols which are in BAD_SYMBOLS_RE from text
    return text

data['summaries']=data['summaries'].apply(clean_text)
data.summaries = data.summaries.replace('\s+', ' ', regex=True)
```

```python
tech = pd.read_csv("Tech_pydictionary_2.csv")

tech=tech.iloc[:,1:2]
tech.head()
```

```python
summary=data
```

```python
tech['Tech_word'] = tech['Tech_word'].astype(str)
all_tech_words = list(tech['Tech_word'].str.lower())
summary["summaries"]=summary["summaries"].astype(str)
all_strings = list(summary["summaries"].str.lower())
```

```python
len(all_tech_words),len(all_strings) 
```

```python
i=1
tech_keys=[]
for item in all_strings:
#     print("***","TEST No,",i,"***")
    i=i+1
    tech_row=[]
    for k in all_tech_words:
        if k in item and len(k)>2:
            tech_row.append(k)
    tech_keys.append(tech_row)
# end = time.time()-start_time # where search ends
# print(end)
```

```python
summary['keywords']=tech_keys
summary.head()
```

```python
import re
def reg_match(txt, keyw):
    sep='|'
    temp = sep.join(['\\b' + i+'\\b' for i in keyw])
    matches = re.findall(temp,txt)
    return list(set(matches))
```

```python
summary['string_matched'] = summary.apply(lambda x: reg_match(x["summaries"], x["keywords"]), axis = 1)
```
